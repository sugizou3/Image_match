```python
def get_contorf(data,per_range= [10 ,99],abs_range=None,num_color=20):
  min_div, max_div = np.percentile(data, per_range)
  if abs_range:
    min_div, max_div = abs_range

  interval_color = np.round(np.linspace(min_div,max_div,num_color),3)
  xArange = np.arange(rd.div.shape[1])*rd.data_span-rd.depthCenter
  yArange = np.arange(rd.div.shape[0])*rd.data_span
  X, Y = np.meshgrid(xArange, yArange)

  return X, Y, data,interval_color
```
