```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
import tqdm
```

```python
kigyou_infos = [] 
for i in  tqdm.tqdm(range(700)):
    try:
        url = 'https://en-hyouban.com/search/industry/aitei-tsushin/?page={}'.format(i)
        # ウェブページを取得する
        response = requests.get(url)

        # レスポンスのテキストをパースする
        soup = BeautifulSoup(response.text, 'html.parser')

        # 年収情報を含む要素を検索
        # ここでは、例として年収情報が<some-element>タグ内にあると仮定します
        kigyou_list = soup.find_all('div', class_='card direct-chat direct-chat-primary result-item style-02 border-0 rounded-0') 
        for kigyou in kigyou_list:
            name = kigyou.find('a',attrs={'class':'company_name_anchor'}).text
            id = kigyou.find('a',attrs={'class':'company_name_anchor'}).get("href").split('/')[-2] 
            salary = float(kigyou.find('div',attrs={'class':'d-flex align-items-center style-01'}).find('span',attrs={'class':'font-weight-bold color-link mb-0 mx-1 f-company-info'}).text)
            if salary > 500:
                kigyou_info = {'name':name,'id':id,'salary':salary}
                kigyou_infos.append(kigyou_info)
            
        time.sleep(1)
    except:
        continue
df_kigyou_info = pd.DataFrame(kigyou_infos)
```

```python
df_kigyou_info.to_pickle('./kigyou_aitei-tsushin.pickle')
```
